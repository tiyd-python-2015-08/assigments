# Vectors

A _vector_ is a series of numbers, usually written like this:

$$
\begin{align}
v = \begin{bmatrix}5 \\ 0\end{bmatrix} \\
w = \begin{bmatrix}3 \\ 4\end{bmatrix}
\end{align}
$$

although we can write it horizontally as well:

$$
\begin{align}
v = \begin{bmatrix}5 & 0\end{bmatrix} \\
w = \begin{bmatrix}3 & 4\end{bmatrix}
\end{align}
$$

It shows motion in a coordinate space, which is not super-important for what you're learning here, but just to note: a two-element vector is motion on a 2-dimensional plane, with the first number being the movement on the X axis and the second number being the movement on the Y axis. We can visualize this on a grid.

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt

def draw_vector(x, y, start_x=0, start_y=0, **kwargs):
    args = dict(head_width=0.15, head_length=0.3, 
                overhang=0.5, length_includes_head=True)
    args.update(kwargs)
    plt.arrow(start_x, start_y, start_x + x, start_y + y, **args)

plt.figure(figsize=(7, 7))

draw_vector(5, 0, color="blue")
draw_vector(3, 4, color="red")

plt.grid(True)
plt.axis([-1, 7, -1, 7])

plt.text(2.5, 0.3, r"$v = \left[\stackrel{5}{0}\right]$", fontsize=20)
plt.text(1.5, 2.3, r"$w = \left[\stackrel{3}{4}\right]$", fontsize=20)
plt.show()
```

For what we will be doing with data, vectors come in very handy, but you need a new style of math to work with them.

## Addition and subtraction

Adding two vectors of the same size is easy:

$$\
\begin{bmatrix}v_1 \\ v_2 \end{bmatrix} +
\begin{bmatrix}w_1 \\ w_2 \end{bmatrix} =
\begin{bmatrix}v_1 + w_1 \\ v_2 + w_2 \end{bmatrix}
$$

as is subtraction:

$$\
\begin{bmatrix}v_1 \\ v_2 \end{bmatrix} -
\begin{bmatrix}w_1 \\ w_2 \end{bmatrix} =
\begin{bmatrix}v_1 - w_1 \\ v_2 - w_2 \end{bmatrix}
$$

```python
plt.figure(figsize=(6, 6))

draw_vector(5, 0, color="blue")
draw_vector(3, 4, color="red")
draw_vector(8, 4, color="purple", linestyle="dotted")

plt.grid(True)
plt.axis([-1, 9, -1, 9])

plt.text(2.5, 0.2, r"$v$", color="blue", fontsize=20)
plt.text(1.5, 2.6, r"$w$", color="red", fontsize=20)
plt.text(4, 2, r"$v + w$", color="purple", fontsize=20)
plt.show()
```

## Multiplication by a scalar

A _scalar_ is a single number. You can multiply a vector by a scalar.

$$
\begin{bmatrix}v_1 \\ v_2 \\ v_3\end{bmatrix} \times
c =
\begin{bmatrix}v_1 \times c \\ v_2 \times c \\ v_3 \times c\end{bmatrix}
$$

## Mean of vectors

With the ability to add vectors and multiply by a scalar, we can find the mean of several vectors.

$$
mean = 
\left(
\begin{bmatrix}v_{1_1} \\ v_{1_2} \\ v_{1_3}\end{bmatrix} +
\begin{bmatrix}v_{2_1} \\ v_{2_2} \\ v_{2_3}\end{bmatrix} +
\ldots +
\begin{bmatrix}v_{n_1} \\ v_{n_2} \\ v_{n_3}\end{bmatrix} +
\right)
\times
\frac{1}{n}
$$

```python
plt.figure(figsize=(7, 7))

draw_vector(5, 0, color="blue")
draw_vector(3, 4, color="red")
draw_vector(1, 5, color="green")
draw_vector(3, 3, color="purple")

plt.grid(True)
plt.axis([-1, 7, -1, 7])

plt.text(1.5, 1.4, r"mean", color="purple", fontsize=20)
plt.show()
```

## Dot Product

The _dot product_ is hard to describe geometrically. You may hear it called the length of the projection of _v_ onto _w_.

Dot products are used a lot in prediction and analysis.

The formula is easy:

$$
v \cdot w = \left(v_1 w_1 + v_2 w_2 + \ldots + v_n w_n\right)
$$


This formula helps us figure out our next and last vector formula.

## Magnitude

Magnitude is the length of a vector. Its formula is $\sqrt{v \cdot v}$.

# Matrices

A _matrix_ is a two-dimensional collection of numbers. You can think of it as a collection of vectors.

$$
A = 
\begin{bmatrix}
2 & 9 & 4 \\
7 & 5 & 3
\end{bmatrix}
$$

Variables for matrices generally are in uppercase.

The _shape_ of a matrix is its number of rows and columns, always in that order. The number of rows is usually called _n_ and the number of columns called _k_. The matrix _A_ above is a 2 x 3 matrix.

## Addition and Subtraction

Matrix addition and subtraction works like vector addition and subtraction.

$$
\begin{bmatrix}
A_{1,1} & A_{1,2} \\
A_{2,1} & A_{2,2} \\
A_{3,1} & A_{3,2}
\end{bmatrix}
+
\begin{bmatrix}
B_{1,1} & B_{1,2} \\
B_{2,1} & B_{2,2} \\
B_{3,1} & B_{3,2}
\end{bmatrix}
=
\begin{bmatrix}
A_{1,1} + B_{1,1} & A_{1,2} + B_{1,2} \\
A_{2,1} + B_{2,1} & A_{2,2} + B_{2,2} \\
A_{3,1} + B_{3,1} & A_{3,2} + B_{3,2} \\
\end{bmatrix}
$$

## Multiplication by a scalar

This also works the same way it did with a vector.

$$
\begin{bmatrix}
A_{1,1} & A_{1,2} & A_{1,3} \\ 
A_{2,1} & A_{2,2} & A_{2, 3} 
\end{bmatrix} 
\times
c =
\begin{bmatrix}
A_{1,1} \times c & A_{1,2} \times c & A_{1,3} \times c \\ 
A_{2,1} \times c & A_{2,2} \times c & A_{2, 3} \times c 
\end{bmatrix} 
$$

## Multiplication by a vector

When you multiply a matrix by a vector, you get a new vector.

In order to multiply a matrix by a vector, you must obey this _shape rule_:

*Shape Rule: The number of rows of the vector must equal the number of columns of the matrix.*

$$
\begin{bmatrix}
A_{1,1} & A_{1,2} \\
A_{2,1} & A_{2,2} \\
A_{3,1} & A_{3,2}
\end{bmatrix}
\times
\begin{bmatrix}
v_1 \\
v_2
\end{bmatrix}
=
\begin{bmatrix}
A_{1,1} \times v_1 + A_{1,2} \times v_2 \\
A_{2,1} \times v_1 + A_{2,2} \times v_2 \\
A_{3,1} \times v_1 + A_{3,2} \times v_2 \\
\end{bmatrix}
$$

or, maybe more clearly:

$$
\begin{bmatrix}
a & b \\
c & d \\
e & f
\end{bmatrix}
\times
\begin{bmatrix}
m \\
n
\end{bmatrix}
=
\begin{bmatrix}
am + bn \\
cm + dn \\
em + fn \\
\end{bmatrix}
$$

Example:

$$
\begin{bmatrix}
3 & 2 \\
4 & 1 \\
0 & 4
\end{bmatrix}
\times
\begin{bmatrix}
2 \\
3
\end{bmatrix}
=
\begin{bmatrix}
3 \times 2 + 2 \times 3 \\
4 \times 2 + 1 \times 3 \\
0 \times 2 + 4 \times 3
\end{bmatrix}
=
\begin{bmatrix}
12 \\
11 \\
12
\end{bmatrix}
$$

## Multiplication by another matrix

When you multiply a matrix by a matrix, you get a new matrix.

*Shape Rule: The number of columns of the first matrix must equal the number of rows of the second matrix.*

$$
\begin{bmatrix}
A_{1,1} & A_{1,2} \\
A_{2,1} & A_{2,2} \\
A_{3,1} & A_{3,2}
\end{bmatrix}
\times
\begin{bmatrix}
B_{1,1} & B_{1,2} \\
B_{2,1} & B_{2,2}
\end{bmatrix}
=
\begin{bmatrix}
A_{1,1}B_{1,1} + A_{1,2}B_{2,1} & A_{1,1}B_{1,2} + A_{1,2}B_{2,2} \\
A_{2,1}B_{1,1} + A_{2,2}B_{2,1} & A_{2,1}B_{1,2} + A_{2,2}B_{2,2} \\
A_{3,1}B_{1,1} + A_{3,2}B_{2,1} & A_{3,1}B_{1,2} + A_{3,2}B_{2,2} \\
\end{bmatrix}
$$

or, maybe more clearly:

$$
\begin{bmatrix}
a & b \\
c & d \\
e & f
\end{bmatrix}
\times
\begin{bmatrix}
w & x \\
y & z
\end{bmatrix}
=
\begin{bmatrix}
aw + by & ax + bz \\
cw + dy & cx + dz \\
ew + fy & ex + fz \\
\end{bmatrix}
$$

Example:

$$
\begin{bmatrix}
3 & 2 \\
4 & 1 \\
0 & 4
\end{bmatrix}
\times
\begin{bmatrix}
2 & 1 \\
3 & 0
\end{bmatrix}
=
\begin{bmatrix}
3 \times 2 + 2 \times 3 & 3 \times 1 + 2 \times 0 \\
4 \times 2 + 1 \times 3 & 4 \times 1 + 1 \times 0 \\
0 \times 2 + 4 \times 3 & 0 \times 1 + 4 \times 0
\end{bmatrix}
=
\begin{bmatrix}
12 & 3 \\
11 & 4 \\
12 & 0
\end{bmatrix}
$$
